```python
import torch
import torch.nn as nn
```

```python
linear = nn.Linear(in_features=4096 , out_features=10)
conv = nn.Conv2d(in_channels=3, out_channels=9, kernel_size=3,stride=1,padding=1)
relu = nn.ReLU(False)
```

```python
X = torch.randn(4096)
y = linear(X)
y.size()
```

```python
X = torch.randn((1,3,7,7))
y = conv(X)
z = relu(y)
```

```python
y.size()
```
